```python
data = [[float(i) for i in l.split()] for l in open("lat_const_CT/data.txt",'r').readlines()[17:25]]
```

```python
plt.plot(lat_const, energy)
font1 = {'family':'serif','color':'Black','size':12}
font2 = {'family':'serif','color':'Black','size':11}
# plt.title("MoS2 - Convergence w.r.t # kpts",fontdict=font1)
plt.xlabel("lattice constant (Bohr)",fontdict = font2)
plt.ylabel("Total Energy (Ry)",fontdict = font2)
plt.ticklabel_format(useOffset=False)
plt.show()
plt.savefig('MoS2_CT_alat.png')
```
