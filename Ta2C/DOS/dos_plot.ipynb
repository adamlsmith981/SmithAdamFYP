```python
import matplotlib.pyplot as plt
from matplotlib import rcParamsDefault
import numpy as np
%matplotlib inline
```

```python

# load data
energy, dos, idos = np.loadtxt('/Users/adamsmith/Documents/GitHub/PH30036_AdamSmith/FYP/Ta2C/DOS/DOS_results/Ta2C.dos.dat', unpack=True)


```

```python
fermiE=-0.203
# make plot
plt.figure(figsize = (12, 6))
plt.plot(energy, dos, linewidth=0.75, color='red')
plt.yticks([])
plt.xlabel('Energy (eV)')
plt.ylabel('DOS')
plt.axvline(x=fermiE, linewidth=0.5, color='k', linestyle=(0, (8, 10)))
plt.xlim(-10, 7.5)
plt.ylim(0, )
plt.fill_between(energy, 0, dos, where=(energy < fermiE), facecolor='red', alpha=0.25)
plt.text(-0.3+fermiE, 10, 'Fermi energy', fontsize='medium', rotation=90)
plt.show()
```
