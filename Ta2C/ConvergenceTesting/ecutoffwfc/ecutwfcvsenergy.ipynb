```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_6 = [[float(i) for i in l.split()] for l in open("ecutwfc_CT/data.txt",'r').readlines()[4:]]
```

```python
ecutwfc1, energy1 = zip(*data_6)
```

```python
plt.plot(ecutwfc1, energy1)
font1 = {'family':'serif','color':'blue','size':12}
font2 = {'family':'serif','color':'darkred','size':10}
plt.title("Convergence test - varying kinectic energy cut off",fontdict = font1)
plt.xlabel("ecutwfc Ry / eV",fontdict = font2)
plt.ylabel("Total Energy Ry / eV",fontdict = font2)
plt.show()
```

```python
plt.plot(ecutwfc1, energy1)
font1 = {'family':'serif','color':'Black','size':12}
font2 = {'family':'serif','color':'Black','size':11}
plt.title("Ta2C - varying K.E cut off",fontdict = font1)
plt.xlabel("ecutwfc Ry",fontdict = font2)
plt.ylabel("Total Energy Ry",fontdict = font2)
plt.ticklabel_format(useOffset=False)
plt.show()
plt.savefig('Ta2C_CT_ecut.png')
```
