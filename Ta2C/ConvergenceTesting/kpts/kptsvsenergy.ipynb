```python
data = [[float(i) for i in l.split()] for l in open("kpts_CT/data.txt",'r').readlines()[4:]]
```

```python
n, k, energy = zip(*data)
```

```python
plt.plot(k, energy)
font1 = {'family':'serif','color':'Black','size':12}
font2 = {'family':'serif','color':'Black','size':11}
plt.title("MoS2 - Convergence w.r.t # kpts",fontdict=font1)
plt.xlabel("Number of K points",fontdict = font2)
plt.ylabel("Total Energy R_y / eV",fontdict = font2)
plt.ticklabel_format(useOffset=False)
plt.show()
plt.savefig('MoS2_CT_kpts.png')
```
